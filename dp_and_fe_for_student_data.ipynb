##จะเป็นการปรับชื่อคณะให้เป็นมาตรฐานเดียวกัน เช่น รวมชื่อที่หมายถึง “วิทยาศาสตร์” หรือ “ศิลปศาสตร์” ให้เป็นชื่อเดียวกัน เพื่อให้ง่ายต่อการวิเคราะห์ข้อมูลและลดความซ้ำซ้อนของชื่อคณะที่เขียนต่างกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name
```

##เพิ่มคอลัมน์ใหม่ชื่อ ‘คณะ_มาตรฐาน’ โดยนำค่าคณะเดิมมาแปลงผ่านฟังก์ชัน standardize_faculty เพื่อให้ชื่อคณะในข้อมูลมีรูปแบบที่เป็นมาตรฐานเดียวกัน



```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

##คำนวณ Q1 กับ Q3 ของข้อมูลอายุ เพื่อดูช่วงปกติของข้อมูลไว้ใช้หาค่าผิดปกติ

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

##ค่าที่ได้จาก Q1 และ Q3 คือ 20.0 และ 22.0 หมายความว่า 25% แรกของข้อมูลอายุจะอยู่ต่ำกว่า 20 และ 75% ของข้อมูลจะอยู่ต่ำกว่า 22 โดยค่าทั้งสองนี้ใช้เป็นขอบเขตในการวิเคราะห์ช่วงข้อมูลปกติ และตรวจหาค่าที่อยู่นอกช่วง

```python
Q1, Q3
```

##หาช่วงของข้อมูลปกติ (IQR) โดยเอา Q3 ลบ Q1 เอาไว้ใช้หาค่าที่หลุดออกจากช่วงปกติ

```python
IQR = Q3 - Q1
```

##แสดงค่า IQR ที่คำนวณได้ เพื่อดูว่าข้อมูลอายุปกติควรอยู่ในช่วงแค่ไหน และเตรียมใช้หาค่าที่หลุดจากช่วงนี้



```python
IQR
```

##จะเป็นการหาช่วงอายุที่ปกติด้วยการใช้ IQR ที่คำนวณไว้แล้ว แล้วเอาไปลบหรือบวกกับ Q1/Q3 ด้วยสูตร 1.5 × IQR

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

##แสดงผลค่าที่ได้จากการคำนวณออกมา

```python
print(lower_bound)
print(upper_bound)
```

##ทำการเลือกเอาข้อมูลเฉพาะแถวที่มีค่าอายุอยู่ในช่วงปกติระหว่าง lower_bound กับ upper_bound เพื่อเอาค่าผิดปกติออก

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

##ดูข้อมูลหลังกรองเอาค่าผิดปกติออกเรียบร้อยแล้ว

```python
df_no_outliers
```

##เช็คแต่ละแถวว่าอายุมากกว่าหรือเท่ากับค่าขอบเขตล่าง (lower_bound) หรือไม่ เพื่อใช้กรองข้อมูลต่อไป

```python
df['อายุ'] >= lower_bound
```

##สร้างตารางข้อมูลชื่อ ส่วนสูง และน้ำหนักนักศึกษา เพื่อเอาไว้ทำวิเคราะห์สุขภาพ

```python
student_health = pd.DataFrame({
    'ชื่อ': ['สมชาย', 'สมหญิง', 'สมศักดิ์'],
    'ส่วนสูง_cm': [170, 165, 175],
    'น้ำหนัก_kg': [65, 55, 80]
})
```

##หาค่า BMI จากน้ำหนักและส่วนสูง แล้วแบ่งกลุ่มสุขภาพตามค่า BMI ว่าอยู่ในช่วงผอม ปกติ อ้วน หรืออ้วนมาก

```python
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

##แสดงข้อมูลสุขภาพนักศึกษาพร้อมค่า BMI และการแบ่งกลุ่มสุขภาพตามเกณฑ์ BMI

```python
student_health
```
